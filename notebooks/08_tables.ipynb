```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Table I: Overall Metrics
fig1, ax1 = plt.subplots(figsize=(10, 2.5))
ax1.axis('off')

cols_summary = ['Model', 'Weighted F1', 'Macro F1']
rows_summary = [
    ['TF-IDF + LogReg',  '0.595', '0.564'],
    ['RoBERTa (100k)',   '0.605', '0.584'],
    ['GPT-4.1 Zero-Shot','0.757', '0.749'],
]

t1 = ax1.table(
    cellText=rows_summary,
    colLabels=cols_summary,
    cellLoc='center',
    loc='center'
)
t1.auto_set_font_size(False)
t1.set_fontsize(11)
t1.scale(1.2, 2.2)

for j in range(len(cols_summary)):
    t1[0, j].set_facecolor('#2d6a9f')
    t1[0, j].set_text_props(color='white', fontweight='bold')

for i in range(1, len(rows_summary) + 1):
    for j in range(len(cols_summary)):
        if i % 2 == 0:
            t1[i, j].set_facecolor('#f7f7f7')
        else:
            t1[i, j].set_facecolor('white')
    t1[i, 0].set_text_props(fontweight='bold')

# Highlight best
for row_idx in [3]:
    for col_idx in [1, 2]:
        t1[row_idx, col_idx].set_facecolor('#d4edda')
        t1[row_idx, col_idx].set_text_props(fontweight='bold')

plt.title('Table I: Model Comparison on Manually Labeled Validation Set',
          fontsize=14, fontweight='bold', pad=10)
plt.tight_layout()
plt.savefig('../figures/table1_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

# Table II: Per-Aspect F1
fig2, ax2 = plt.subplots(figsize=(14, 4.5))
ax2.axis('off')

cols_aspect = ['Aspect',
               'TF-IDF\nPos', 'TF-IDF\nNeg',
               'RoBERTa\nPos', 'RoBERTa\nNeg',
               'GPT-4.1\nZero-Shot Pos', 'GPT-4.1\nZero-Shot Neg']

rows_aspect = [
    ['Product Quality','0.65', '0.56', '0.66', '0.56', '0.82', '0.77'],
    ['Service',        '0.71', '0.74', '0.74', '0.76', '0.91', '0.87'],
    ['Wait Time',      '0.44', '0.55', '0.53', '0.58', '0.62', '0.72'],
    ['Price Value',    '0.59', '0.66', '0.69', '0.66', '0.85', '0.82'],
    ['Cleanliness',    '0.71', '0.55', '0.76', '0.58', '0.91', '0.77'],
    ['Atmosphere',     '0.48', '0.22', '0.47', '0.21', '0.70', '0.60'],
    ['General',        '0.51', '0.52', '0.53', '0.46', '0.60', '0.53'],
]

t2 = ax2.table(
    cellText=rows_aspect,
    colLabels=cols_aspect,
    cellLoc='center',
    loc='center'
)
t2.auto_set_font_size(False)
t2.set_fontsize(10)
t2.scale(1.2, 2.0)

for j in range(len(cols_aspect)):
    t2[0, j].set_facecolor('#2d6a9f')
    t2[0, j].set_text_props(color='white', fontweight='bold')

for i in range(1, len(rows_aspect) + 1):
    for j in range(len(cols_aspect)):
        if i % 2 == 0:
            t2[i, j].set_facecolor('#f7f7f7')
        else:
            t2[i, j].set_facecolor('white')
    t2[i, 0].set_text_props(fontweight='bold')

for row_idx in range(1, len(rows_aspect) + 1):
    for col_idx in [5, 6]:
        t2[row_idx, col_idx].set_facecolor('#d4edda')
        t2[row_idx, col_idx].set_text_props(fontweight='bold')

plt.title('Table II: Per-Aspect F1 on Manually Labeled Validation Set',
          fontsize=20, fontweight='bold', pad=5)
plt.tight_layout()
plt.savefig('../figures/table2_aspect_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```
